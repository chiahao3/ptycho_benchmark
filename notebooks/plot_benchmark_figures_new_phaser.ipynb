# Clean up DataFrame with clean iter time and compilation time (mean, std)

```python
# Define the packages that actually have a JIT/compilation overhead
compiled_packages = ["ptyrad_default", "ptyrad_max", "ptyrad_reduce", "phaser", "phaser_buff_no_unroll_10", "phaser_buff_8_unroll_10", "phaser_buff_128_unroll_10"]

# Select the successful runs and OOM runs
df_success = df[df["status"] == "success"].copy()
df_oom = df[df["status"] == "OOM"].copy()

# Filter out the 1st iteration that contains significant warm up and compilation time
WARMUP = 1 

# ==========================================
# 1. CALCULATE METRICS ON SUCCESSFUL RUNS
# ==========================================
# Calculate Steady State (clean_iter_time)
df_success["clean_iter_time"] = df_success["iter_times"].apply(
    lambda x: np.mean(x[WARMUP:]) if isinstance(x, list) and len(x) > WARMUP else (x[0] if isinstance(x, list) and len(x) > 0 else np.nan)
)

# Calculate Compilation Time (First Iteration - Steady State)
def calc_compile_time(row):
    if row["label"] in compiled_packages:
        x = row["iter_times"]
        if isinstance(x, list) and len(x) > 0:
            first_iter = x[0]
            steady_state = row["clean_iter_time"]
            
            # Ensure we have a valid steady state to subtract from
            if pd.notnull(steady_state):
                return first_iter - steady_state
    
    # Returns NaN if not in the list, or if the data is corrupted
    return np.nan

df_success["compile_time"] = df_success.apply(calc_compile_time, axis=1)

# ==========================================
# 2. PREP OOMs AND MERGE
# ==========================================
# Explicitly set both metrics to NaN for OOMs so they don't break the math
df_oom["clean_iter_time"] = np.nan
df_oom["compile_time"] = np.nan

# Merge them together BEFORE the groupby
df_success_and_oom = pd.concat([df_success, df_oom], ignore_index=True)

# ==========================================
# 3. AGGREGATE ACROSS ROUNDS
# ==========================================
# Group by the master dataframe, calculating mean and std for BOTH metrics
agg_df = df_success_and_oom.groupby(["label", "device", "batch", "pmode", "slice"]).agg({
    "clean_iter_time": ['mean', 'std'],
    "compile_time": ['mean', 'std']
}).reset_index()

# Flatten the annoying MultiIndex columns that Pandas creates when aggregating multiple columns
# This turns ('clean_iter_time', 'mean') into 'clean_iter_time_mean'
agg_df.columns = ['_'.join(col).strip('_') for col in agg_df.columns.values]
```

# Plot the Table figure

```python
# ==========================================
# 0. DEFINE YOUR PLOT SUBSET
# ==========================================
# Tweak these lists to plot exactly what you want. 
# The figure size and grid will automatically adapt!
plot_packages = ["phaser",
                 "phaser_buff_no_unroll_10",
                "phaser_buff_8_unroll_10" ,
                 "phaser_buff_128_unroll_10", 
 
                 "ptyrad_max", 
                 "ptyrad_reduce", 
                 "ptyrad_default", 
                 "ptyrad",
                #  "quantem",
                #  "py4dstem", 
                #  "ptyshv"
                 ]
plot_batches = [
                4,
                8,
                16,
                32,
                64,
                128,
                256,
                512,
                1024]
plot_slices = [1, 3, 6, 12]
plot_pmodes = [1, 3, 6, 12] 

# ==========================================
# 1. PREP THE DATA FOR TABLE
# ==========================================

show_std = False # Whether to show std with the iter time
empty_value = " — " # The value used to fill configurations that were never run (i.e., like PtyShv and py4DSTEM at extremely small batch size or large slices because it'd be too slow)

# Because df_oom had NaN times, their 'mean' is now NaN. 
# The lambda function catches these and explicitly tags them as "OOM".
if show_std:
    agg_df['time_str'] = agg_df.apply(
        lambda r: f"{r['clean_iter_time_mean']:.3f} ± {r['clean_iter_time_std']:.3f}" if pd.notnull(r['clean_iter_time_mean']) else "OOM", 
        axis=1
    )
else:
    agg_df['time_str'] = agg_df.apply(
        lambda r: f"{r['clean_iter_time_mean']:.3f}" if pd.notnull(r['clean_iter_time_mean']) else "OOM", 
        axis=1
    )

# Filter the dataframe down to ONLY the subset we defined above
agg_df = agg_df[
    agg_df['label'].isin(plot_packages) &
    agg_df['batch'].isin(plot_batches) &
    agg_df['slice'].isin(plot_slices) &
    agg_df['pmode'].isin(plot_pmodes)
]

# ==========================================
# 2. DYNAMIC PLOTTING SETUP
# ==========================================
package_colors = {
    "phaser": "#cfe2f3",   # light blue
    "phaser_buff_no_unroll_10": "#e9d7a4",   # light yellow
    "phaser_buff_8_unroll_10": "#d0e0e3", # cyan
    "phaser_buff_128_unroll_10": "#f4cccc", # coral

    "ptyrad_max":"#fce5cd", # light orange
    "ptyrad_reduce": "#d9ead3",   # light green
    "ptyrad_default": "#E6BDBD", # light red/pink
    "ptyrad": "#D6BDE6", # light purple
    # "quantem": "#e9d7a4",   # light yellow
    # "py4dstem": "#d0e0e3", # cyan
    # "ptyshv": "#f4cccc", # coral
}
probe_header_color = "#E4E4E4"

def parse_val(val_str):
    """ Parsing the values for table cell styling """
    val_str = str(val_str)
    # Catch both OOM and NaN so they don't break the min() function while styling
    if "OOM" in val_str or "NaN" in val_str or empty_value in val_str:
        return float('inf')
    try:
        return float(val_str.split('±')[0].strip())
    except:
        return float('inf')

# Calculate dynamic figsize based on the number of elements requested
# (Adjust these multipliers if your text feels too cramped or too loose)
col_width = 2.55 * len(plot_pmodes)    
row_height = 0.35 * len(plot_packages)


fig_width = (col_width * len(plot_slices))
fig_height = (row_height * len(plot_batches))

width_ratios = [0.1] + [1] * len(plot_slices)
height_ratios = [0.05] + [1] * len(plot_batches)

fig = plt.figure(figsize=(fig_width, fig_height), dpi=100)
gs = gridspec.GridSpec(
    nrows=len(plot_batches) + 1,
    ncols=len(plot_slices) + 1,
    figure=fig,
    wspace=0.10,
    hspace=0.10,
    width_ratios=width_ratios, 
    height_ratios=height_ratios
)

# Headers
for j, s in enumerate(plot_slices):
    ax = fig.add_subplot(gs[0, j+1])
    ax.axis("off")
    ax.text(0.4, 0, f"{s} slice{'s' if s > 1 else ''}",
            ha='center', va='center', fontsize=12, fontweight='bold')

for i, b in enumerate(plot_batches):
    ax = fig.add_subplot(gs[i+1, 0])
    ax.axis("off")
    ax.text(0, 0.5, f"batch {b}",
            ha='center', va='center', fontsize=12, rotation=90, fontweight='bold')

# ==========================================
# 3. BUILD THE MAIN GRID
# ==========================================
for i, b in enumerate(plot_batches):
    for j, s in enumerate(plot_slices):
        ax = fig.add_subplot(gs[i+1, j+1])
        ax.axis("off")

        subset = agg_df[(agg_df["batch"] == b) & (agg_df["slice"] == s)].copy()
        pivot = subset.pivot(index="label", columns="pmode", values="time_str")
        
        # 1. Force the exact structure you requested in Section 0
        pivot = pivot.reindex(index=plot_packages, columns=plot_pmodes)
        
        # 2. Fill missing configurations (things that were never run) with empty_value
        pivot = pivot.fillna(empty_value)

        formatted_probes = [f"{int(p)} probe{'s' if p > 1 else ''}" for p in pivot.columns]
        table_data = [formatted_probes] + pivot.values.tolist()
        row_labels = list(pivot.index)

        table = ax.table(
            cellText=table_data,
            rowLabels=[""] + row_labels,
            colLabels=None,
            cellLoc='center',
            loc='center',
            bbox=[0.15, 0, 0.75, 0.95]
        )
        table.auto_set_font_size(False)
        table.set_fontsize(10)

        # Styling
        for col in range(len(formatted_probes)):
            table[0, col].set_facecolor(probe_header_color)

        for row_idx, pkg in enumerate(row_labels):
            table[row_idx + 1, -1].set_facecolor(package_colors.get(pkg, '#eeeeee'))

        # Bold best values
        for col_idx, col_name in enumerate(pivot.columns):
            values = pivot[col_name].tolist()
            parsed = [parse_val(v) for v in values]
            
            if all(v == float('inf') for v in parsed):
                continue 
                
            min_val = min(parsed)
            for row_idx, pkg in enumerate(row_labels):
                cell_val = str(pivot.iloc[row_idx, col_idx])
                cell = table[row_idx + 1, col_idx]
                
                # Ignore both explicit OOMs and empty values
                if "OOM" in cell_val or empty_value in cell_val:
                    continue
                if parsed[row_idx] == min_val:
                    cell.set_text_props(weight='bold', color='black')
                    cell.set_facecolor(package_colors.get(pkg, '#dddddd'))

# ==========================================
# 4. TITLES & EXPORT
# ==========================================
fig.suptitle(f"Iteration Time (sec) Benchmark plotted at {date_str}", fontsize=20, y=0.98)
fig.text(0.5, 0.95, "Measured on a 20 GB MIG slice from an 80 GB NVidia A100 using the 128x128 tBL-WSe2 dataset.", 
         ha='center', va='top', fontsize=12)

# This physically pushes the top of the GridSpec down to 90% of the figure height,
# leaving the top 10% completely safe for your titles, without triggering the warning!
plt.subplots_adjust(top=0.93, bottom=0.05, left=0.05, right=0.95)


fig_name = f"iter_time_table_{date_str}"
plt.savefig(f"{output_dir}/{fig_name}.png", bbox_inches="tight")

plt.show()
```

# Plot the compilation and iter time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ==========================================
# 0. DEFINE YOUR PLOT SUBSET
# ==========================================
# Keep this tight to tell a specific story in your paper/presentation
plot_packages = [
                "phaser",
                "phaser_buff_no_unroll_10",
                 "phaser_buff_8_unroll_10",
                 "phaser_buff_128_unroll_10",
                "ptyrad_max", 
                "ptyrad_reduce", 
                "ptyrad_default", 
                 "ptyrad", 
                #  "quantem"
                 ]
plot_batches = [4, 16, 64, 256]
plot_slices = [6, 12]
target_pmode = 6  # Isolate one probe mode for visual clarity

# True: Print the total time (Iter + Compile) at the end of every bar
# False: Keep the bars clean and visually driven
show_values = False

# The Toggle Switch
# True: All subplots share the exact same X-axis scale (great for global trends)
# False: Each subplot auto-scales to fit its own data (great for zooming in on tiny values)
global_x_scale = False

# Filter the dataframe
plot_df = agg_df[
    agg_df['label'].isin(plot_packages) &
    agg_df['batch'].isin(plot_batches) &
    agg_df['slice'].isin(plot_slices) &
    (agg_df['pmode'] == target_pmode)
].copy()

# Enforce the exact order of packages (reversed so the first package is at the top of the Y-axis)
plot_df['label'] = pd.Categorical(plot_df['label'], categories=plot_packages[::-1], ordered=True)

# Handle NaNs: Packages without compilation time (like py4dstem) need 0 instead of NaN for the math to work
plot_df['compile_time_mean'] = plot_df['compile_time_mean'].fillna(0)

# Calculate the global X limit (Max Iteration + Max Compilation)
if global_x_scale:
    # Fill iter time NaNs with 0 purely for the math so OOMs don't break the max() function
    max_total_time = (plot_df['clean_iter_time_mean'].fillna(0) + plot_df['compile_time_mean']).max()
    global_x_limit = max_total_time * 1.20  # Add a 5% visual padding to the right edge

# ==========================================
# 1. DYNAMIC GRID SETUP
# ==========================================
# Scale the figure based on how many subplots you are rendering
col_width = 3.5 
row_height = 0.5 * len(plot_packages)

fig_width = col_width * len(plot_slices)
fig_height = row_height * len(plot_batches)

fig = plt.figure(figsize=(fig_width, fig_height), dpi=100)
gs = gridspec.GridSpec(
    nrows=len(plot_batches), 
    ncols=len(plot_slices), 
    figure=fig,
    wspace=0.10,  # Tight spacing horizontally
    hspace=0.35   # Room for X-axis labels between rows
)

# ==========================================
# 2. BUILD THE STACKED BARS
# ==========================================
for i, b in enumerate(plot_batches):
    for j, s in enumerate(plot_slices):
        
        ax = fig.add_subplot(gs[i, j])
        
        # Extract and sort data for this specific cell
        cell_data = plot_df[(plot_df["batch"] == b) & (plot_df["slice"] == s)].sort_values('label')
        
        y_positions = np.arange(len(cell_data))
        labels = cell_data['label'].tolist()
        
        compile_times = cell_data['compile_time_mean'].values
        iter_times = cell_data['clean_iter_time_mean'].values
        
        # 1. Plot the Iteration Time FIRST (The Base)
        # Anchoring this at x=0 allows for direct visual comparison of the recurring cost
        iter_colors = [package_colors.get(pkg, '#999999') for pkg in labels]
        ax.barh(
            y_positions, 
            iter_times, 
            color=iter_colors, 
            edgecolor="white",
            label="Steady Iteration" if (i==0 and j==0) else ""
        )

        # 2. Plot the Compile Time SECOND (The Stack)
        # Using np.nan_to_num ensures Matplotlib doesn't panic if iter_times has NaNs (OOMs)
        ax.barh(
            y_positions, 
            compile_times, 
            left=np.nan_to_num(iter_times), 
            color="#e0e0e0", 
            edgecolor="white", 
            hatch="///", 
            label="Compilation Time" if (i==0 and j==0) else ""
        )

        # Apply scaling and padding logic
        if global_x_scale:
            # We add 15% padding globally if we are printing text so it fits
            pad_multiplier = 1.15 if show_values else 1.05
            ax.set_xlim(0, global_x_limit * (pad_multiplier / 1.05))
        else:
            # Calculate the local max for this specific cell
            local_max = np.max(np.nan_to_num(iter_times) + compile_times)
            
            # Add 20% padding locally if printing text so the labels fit inside the box
            pad_multiplier = 1.20 if show_values else 1.05
            if local_max > 0:
                ax.set_xlim(0, local_max * pad_multiplier)
        
        # 3. Handle OOM annotations AND Text Values gracefully
        for idx, (ct, it) in enumerate(zip(compile_times, iter_times)):
            if pd.isna(it):
                # OOM Handler
                ax.text(0.05, idx, "OOM", color='red', va='center', fontsize=9, fontweight='bold')
            elif show_values:
                # Value Printer: Place the text just to the right of the iter time bar
                # Format to 2 decimal places (adjust if you need more precision)
                text_str = f"{it:.2f}s"
                
                # Add a tiny gap (2% of the total axis width) between the bar edge and the text
                x_offset = ax.get_xlim()[1] * 0.02
                ax.text(it + x_offset, idx, text_str, color='black', va='center', fontsize=8)

        # 4. Styling the Cell
        ax.set_yticks(y_positions)
        
        # Only show the package names on the far-left column
        if j == 0:
            ax.set_yticklabels(labels, fontsize=10)
        else:
            ax.set_yticklabels([])
            ax.tick_params(axis='y', length=0)
            
        # Add grid lines behind the bars for easy value reading
        ax.xaxis.grid(True, linestyle='--', alpha=0.5)
        ax.set_axisbelow(True)
        ax.set_xlabel("Iter Time (sec)", fontsize=9)
        
        # Outer headers
        if i == 0:
            ax.set_title(f"{s} Slice{'s' if s > 1 else ''}", fontsize=12, fontweight='bold')
        if j == len(plot_slices) - 1:
            # Place batch size label on the far right edge
            ax.text(1.05, 0.5, f"Batch {b}", transform=ax.transAxes, 
                    va='center', rotation=-90, fontsize=12, fontweight='bold')

# ==========================================
# 3. TITLES, LEGEND & EXPORT
# ==========================================
fig.suptitle(f"First Iteration Total Cost (sec) plotted at {date_str} with {target_pmode} probe modes", fontsize=18, y=1.02)

fig.text(0.5, 1.0, "Measured on a 20 GB MIG slice from an 80 GB NVidia A100 using the 128x128 tBL-WSe2 dataset.", 
         ha='center', va='top', fontsize=12)

# Extract the legend handles from the very first subplot and put them at the top of the figure
handles, lgd_labels = fig.axes[0].get_legend_handles_labels()
fig.legend(handles, lgd_labels, loc='upper center', bbox_to_anchor=(0.5, 0.98), ncol=2, frameon=False, fontsize=11)

plt.subplots_adjust(top=0.93, bottom=0.1, left=0.1, right=0.95)

fig_name = f"compile_and_iter_time_stacked_bars_p{target_pmode}_{date_str}"
plt.savefig(f"{output_dir}/{fig_name}.png", bbox_inches="tight")

plt.show()
```
